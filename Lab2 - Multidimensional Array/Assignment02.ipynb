```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))
```

#Zigzag Walk

```python
def walk_zigzag(floor):
  r,c=floor.shape
  for i in range(c):
    if i%2==0:
      for j in range(r):
        if j%2==0:
          print(floor[j][i],end=' ')
    else:
      for j in range(r-1,-1,-1):
        if j%2!=0:
          print(floor[j][i],end=' ')
    print()


floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0'],
                  ['1' , '4' , '2' , '8' , '6']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9 1
# 1 2
# 4 7 2
# 4 9
# 1 8 6
print('################')
floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9
# 1 2
# 4 7
# 4 9
# 1 8
```

#Wall Up Trost District

```python
def create_fence(district, depth):
    district_row, district_col = district.shape
    tr=district_row+(depth*2)
    tc=district_col+(depth*2)
    new=np.full((tr,tc),8,dtype=int)
    for i in range(district_row):
      for j in range(district_col):
        new[i+depth][j+depth]=district[i][j]

    return new

depth = 1
district = np.array([[2,3,4], [3,4,6], [2,1,4]])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
#This will print
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
# |  8  |  2  |  3  |  4  |  8  |
# -------------------------------
# |  8  |  3  |  4  |  6  |  8  |
# -------------------------------
# |  8  |  2  |  1  |  4  |  8  |
# -------------------------------
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
print('################')
depth = 2
district = np.array([
                 [2,3,4,1],
                 [3,4,6,5],
                 [2,1,4,7]
                ])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
```

#Crows vs Cats

```python
def strength_difference(clubroom):
  strength_diff = np.zeros(len(clubroom)*(len(clubroom)-1)//2, dtype = int)
  idx=0
  for i in range(len(clubroom)):
    for j in range(i,len(clubroom)):
      if i!=j:
        strength_diff[idx]=(clubroom[i][j]-clubroom[j][i])
        idx+=1
  return strength_diff

clubroom = np.array([
[1,  2,  9,  7],
[4,  5,  1,  8],
[3,  6,  2,  7],
[2,  8,  6,  3]
])
print_matrix(clubroom)
returned_value = strength_difference(clubroom)
print('Strength Difference Array is : ', returned_value)
unittest.output_test(returned_value, np.array([-2, 6, 5, -5, 0, 1]))
```

#ATM's Triangle

```python
def atm_triangle(n):
  new=np.full((n,n),None)
  for i in range(n):
    for j in range(i+1):
      if j==0 or j==i:
        new[i][j]=i+1
      else:
        s=0
        for x in range(j+1):
          s+=new[i-1][x]
        new[i][j]=s

  return new

def print_without_none(matrix):
  for i in range(len(matrix)):
    for j in range(len(matrix)):
      if matrix[i][j]!=None:
        print(matrix[i][j],end=' ')
    print()

n = 5
returned_value = atm_triangle(n)
print_without_none(returned_value)
#This should print
# 1
# 2  2
# 3  4  3
# 4  7  10  4
# 5  11 21  25  5
```

#Trace The BOT

```python
def moving_around(cmds):
  grid = np.full((7, 7), '.')
  x,y=3,3
  grid[x,y]='-'
  for i in range(len(cmds)):
    if cmds[i]==1:
      new_x,new_y=x - 2,y - 3
    elif cmds[i]==2:
      new_x,new_y=x - 2,y - 1
    elif cmds[i]==3:
      new_x,new_y=x - 3,y - 2
    elif cmds[i]==4:
      new_x,new_y=x - 1,y - 3
    elif cmds[i]==5:
      new_x,new_y=x - 2,y + 1
    elif cmds[i]==6:
      new_x,new_y=x - 2,y + 3
    elif cmds[i]==7:
      new_x,new_y=x - 3,y + 2
    elif cmds[i]==8:
      new_x,new_y=x - 1,y + 3
    elif cmds[i]==9:
      new_x,new_y=x + 2,y - 3
    elif cmds[i]==10:
      new_x,new_y=x + 2,y - 1
    elif cmds[i]==11:
      new_x,new_y=x + 1,y - 2
    elif cmds[i]==12:
      new_x,new_y=x + 3,y - 1

    if 0 <= new_x < 7 and 0 <= new_y < 7:
      grid[new_x,new_y]='*'
      x,y=new_x,new_y
  grid[x,y]='/'

  return grid

cmds = np.array([5,11,2,9])
result = moving_around(cmds)
print_matrix(result)
#This should print
# -------------------------------------------
# |  .  |  /  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  *  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  *  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  -  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
```
