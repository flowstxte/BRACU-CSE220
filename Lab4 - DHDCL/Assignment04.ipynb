```python
class Patient:

  def __init__(self,id,name,age,bloodgroup):
    self.id=id
    self.name=name
    self.age=age
    self.bloodgroup=bloodgroup
    self.next=None
    self.prev=None

class WRM:

  def __init__(self):
    self.i=Patient(None,None,None,None)
    self.i.next=self.i
    self.i.prev=self.i
    self.head=self.i

  def RegisterPatient(self,id,name,age,bloodgroup):
    p=Patient(id,name,age,bloodgroup)
    self.i.next=p
    p.next=self.head
    p.prev=self.i
    self.head.prev=p
    self.i=p
    print('Success. Patient added.')

  def ServePatient(self):
    if self.head.next==self.head:
      print('No patient in the waiting room.')
    else:
      cp=self.head.next
      print('Serving patient:',cp.name)
      self.head.next=cp.next
      cp.next.prev=self.head
      cp=None

  def CancelAll(self):
    self.head.next=self.head
    self.head.prev=self.head
    self.i=self.head
    print('All patients cancelled. Doctor can go home.')

  def CanDoctorGoHome(self):
    if self.head.next==self.head:
      return True
    else:
      return False

  def ShowAllPatient(self):
    if self.head.next==self.head:
      print('No patient in the waiting room.')
    else:
      i=self.head.next
      while i!=self.head:
        print(f'Patient {i.id}: {i.name}')
        i=i.next

  def ReverseTheLine(self):
    x=self.head.next
    while x!=self.head:
      temp1=x.next
      temp2=x.prev
      x.next=temp2
      x.prev=temp1
      x=x.prev
    temp1=self.head.next
    temp2=self.head.prev
    self.head.next=temp2
    self.head.prev=temp1
    print('Success. Line reversed.')

wrm=WRM()

while True:
  s=input('''a. Add Patient
b. Serve Patient
c. Show All patients
d. Can Doctor go Home?
e. Cancel all Appointment
f. ReverseTheLine
g. Exit
Enter your choice: ''')

  if s=='a':
    id=int(input('Enter ID: '))
    name=input('Enter Name: ')
    age=int(input('Enter Age: '))
    bloodgroup=input('Enter Bloodgroup: ')
    wrm.RegisterPatient(id,name,age,bloodgroup)
  if s=='b':
    wrm.ServePatient()
  if s=='c':
    wrm.ShowAllPatient()
  if s=='d':
    if wrm.CanDoctorGoHome():
      print('Yes. Doctor can go home.')
    else:
      print('No. Doctor can not go home.')
  if s=='e':
    wrm.CancelAll()
  if s=='f':
    wrm.ReverseTheLine()
  if s=='g':
    break
```
